```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')

from skimage.io import imread, imshow
from skimage.transform import resize

import tensorflow as tf
import gc

import pickle

import tensorflow_datasets as tfds
```

```python
dataset, info = tfds.load(name="stanford_dogs", with_info=True)
```

```python
test_data = dataset['test']
IMG_LEN = 299
IMG_SHAPE = (IMG_LEN, IMG_LEN, 3)
N_BREEDS = 60
retrieved_batches = 200
test_x = []
test_y = []
```

```python
i = 0

for x in test_data:
  if i == retrieved_batches:
    break
  if (x['label'].numpy() > N_BREEDS - 1) :
    continue
    
  img = x['image'] / 255
  img = tf.image.resize(img, (IMG_LEN, IMG_LEN), method='nearest')
  test_x.append(np.array(img))
  label = tf.one_hot(np.array(x['label']),N_BREEDS)
  test_y.append(label)
  i += 1
```

```python
test_x = np.array(test_x)
test_y = np.array(test_y)
print(test_x.shape)
print(test_y.shape)
```

```python
model = tf.keras.models.load_model('drive/MyDrive/multi_label_model')
```

```python
opener = open(f'drive/MyDrive/dict_of_labels.pickle', 'rb')
named_labels = pickle.load(opener)
opener.close()
```

```python
results = model.evaluate(test_x, test_y)
```
